```python
from itertools import product

import numpy as np
from sklearn import datasets
from scipy.stats import multivariate_normal as mvn
import matplotlib.pyplot as plt
import random
import pandas as pd

from ripser import Rips
from persim import PersImage
```

```python
cloud1 = [[random.uniform(-1,0), random.uniform(-1,2), random.uniform(-1,1)] for x in range(10)]
cloud2 = [[random.uniform(1,3), random.uniform(-3,-1), random.uniform(-1,1)] for x in range(20)]
cloud3 = [[random.uniform(2,5), random.uniform(0,3), random.uniform(-1,1)] for x in range(12)]

data = np.concatenate([cloud1, cloud2])
```

```python
rips = Rips()
dgms = rips.fit_transform(data)


plt.figure(figsize=(10,5))
plt.subplot(121)
plt.scatter(data[:,0], data[:,1], s=4)
plt.title("Scatter plot of noisy data")

plt.subplot(122)
rips.plot(dgms, legend=False, show=False)
plt.title("Persistence diagram of $H_0$ and $H_1$")
plt.show()
```
